```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from shapely.geometry import LineString

sns.set_style("darkgrid", {"grid.color": ".6"})
```

# Raff's Method

```python
data_df = pd.read_csv('clean_data.csv')
data_df
```

```python
male_accepted_df = data_df[(data_df['Gender'] == 'm') & (data_df['Accepted'] == True)].reset_index(drop=True)
male_rejected_df = data_df[(data_df['Gender'] == 'm') & (data_df['Accepted'] == False)].reset_index(drop=True)
female_accepted_df = data_df[(data_df['Gender'] == 'f') & (data_df['Accepted'] == True)].reset_index(drop=True)
female_rejected_df = data_df[(data_df['Gender'] == 'f') & (data_df['Accepted'] == False)].reset_index(drop=True)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15,8), tight_layout=True)

sns.ecdfplot(data=male_accepted_df, x='Gap', ax=axs[0], palette=["C0"], legend=True)
sns.ecdfplot(data=male_rejected_df, x='Gap', ax=axs[0], palette=["C1"], complementary=True, legend=True)
axs[0].set_title("Raff's Method (Male)", fontsize=20)
axs[0].set_xlabel('Gap size [s]', fontsize=14)
axs[0].set_ylabel('Probability', fontsize=14)
axs[0].legend(labels=['Fa(t)', '1-Fr(t)'])
# legend = axs[0].get_legend()
# legend.set_title('')
# legend.texts[0].set_text('1-Fr(t)')
# legend.texts[1].set_text('Fa(t)')

sns.ecdfplot(data=female_accepted_df, x='Gap', ax=axs[1], palette=["C0"])
sns.ecdfplot(data=female_rejected_df, x='Gap', ax=axs[1], palette=["C1"], complementary=True)
axs[1].set_title("Raff's Method (Female)", fontsize=20)
axs[1].set_xlabel('Gap size [s]', fontsize=14)
axs[1].set_ylabel('Probability', fontsize=14)
axs[1].legend(labels=['Fa(t)', '1-Fr(t)'])

fig.savefig('vis/Raffs_method.png')
```

```python
def raff_calculator(accepted_gaps, rejected_gaps):
    
    '''
    Returns the critical gap as estimated with Raff's method.

            Parameters:
                    accepted_gaps (numpy array or pandas series): data for the accepted gaps as floats
                    accepted_gaps (numpy array or pandas series): data for the rejected gaps as floats

            Returns:
                    gap (float): estimated critical gap, in the same units as the input data.
    '''
    
    # Sort data
    accepted = np.sort(accepted_gaps)
    rejected = np.sort(rejected_gaps)
    
    # Get cumulative probability of accepted gaps, and the complement of the cumulative probability of rejected gaps
    accepted_p = 1. * np.arange(len(accepted)) / (len(accepted) - 1)
    rejected_p = 1 - (1. * np.arange(len(rejected)) / (len(rejected) - 1))
    
    # Turn data into a line
    accepted_line = LineString(np.column_stack((accepted, accepted_p)))
    rejected_line = LineString(np.column_stack((rejected, rejected_p)))
    
    # Find the gap in the intersection of both curves according to Raff's definition
    gap = accepted_line.intersection(rejected_line).x
    return gap
```

```python
male_accepted_data = male_accepted_df['Gap']
male_rejected_data = male_rejected_df['Gap']

gap_male = raff_calculator(male_accepted_data, male_rejected_data)

female_accepted_data = female_accepted_df['Gap']
female_rejected_data = female_rejected_df['Gap']

gap_female = raff_calculator(female_accepted_data, female_rejected_data)

print('The gap for male drivers is ' + "{:.3f}".format(gap_male) +' seconds.')
print()
print('The gap for female drivers is ' + "{:.3f}".format(gap_female) +' seconds.')
```

# Maximum Likelihood Method

```python
male_normal_df = pd.read_csv('ml/male_normal.csv')
male_gamma_df = pd.read_csv('ml/male_gamma.csv')
female_normal_df = pd.read_csv('ml/female_normal.csv')
female_gamma_df = pd.read_csv('ml/female_gamma.csv')
male_normal_df
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10,10), tight_layout=True)
n = 3000 # Arbitrary big number for distribution draws

male_normal_mean = 5.67880303742311 # These values are obtained from the optimization in Excel
male_normal_std = 1.97911493919066
male_normal = np.random.normal(male_normal_mean, male_normal_std, n)
sns.ecdfplot(data=male_normal_df, x='accepted', ax=axs[0][0], color='C0')
sns.ecdfplot(data=male_normal_df, x='rejected', ax=axs[0][0], color='C1')
sns.ecdfplot(x=male_normal, ax=axs[0][0], color='C2')
axs[0][0].legend(labels=['Accepted', 'Rejected', 'Estimated'])
axs[0][0].set_title('Male Normal', fontsize=20)
axs[0][0].set_xlabel('Gap size [s]', fontsize=14)
axs[0][0].set_ylabel('Frequency', fontsize=14)

male_gamma_scale = 8.15670412601199
male_gamma_shape = 0.684896468318653
male_gamma = np.random.gamma(male_gamma_scale, male_gamma_shape, n)
sns.ecdfplot(data=male_gamma_df, x='accepted', ax=axs[0][1], color='C0')
sns.ecdfplot(data=male_gamma_df, x='rejected', ax=axs[0][1], color='C1')
sns.ecdfplot(x=male_normal, ax=axs[0][1], color='C2')
axs[0][1].legend(labels=['Accepted', 'Rejected', 'Estimated'])
axs[0][1].set_title('Male Gamma', fontsize=20)
axs[0][1].set_xlabel('Gap size [s]', fontsize=14)
axs[0][1].set_ylabel('Frequency', fontsize=14)

female_normal_mean = 6.61943641452831
female_normal_std = 1.93479077916473
female_normal = np.random.normal(female_normal_mean, female_normal_std, n)
sns.ecdfplot(data=female_normal_df, x='accepted', ax=axs[1][0], color='C0')
sns.ecdfplot(data=female_normal_df, x='rejected', ax=axs[1][0], color='C1')
sns.ecdfplot(x=female_normal, ax=axs[1][0], color='C2')
axs[1][0].legend(labels=['Accepted', 'Rejected', 'Estimated'])
axs[1][0].set_title('Female Normal', fontsize=20)
axs[1][0].set_xlabel('Gap size [s]', fontsize=14)
axs[1][0].set_ylabel('Frequency', fontsize=14)

female_gamma_scale = 10.7979768888072
female_gamma_shape = 0.613516873287964
female_gamma = np.random.gamma(female_gamma_scale, female_gamma_shape, n)
sns.ecdfplot(data=female_gamma_df, x='accepted', ax=axs[1][1], color='C0')
sns.ecdfplot(data=female_gamma_df, x='rejected', ax=axs[1][1], color='C1')
sns.ecdfplot(x=female_normal, ax=axs[1][1], color='C2')
axs[1][1].legend(labels=['Accepted', 'Rejected', 'Estimated'])
axs[1][1].set_title('Female Gamma', fontsize=20)
axs[1][1].set_xlabel('Gap size [s]', fontsize=14)
axs[1][1].set_ylabel('Frequency', fontsize=14)
```
